```python
#Libraries used for analysis and visualization
import numpy as np   #numpy is used to perform various mathematical operations on arrays
import pandas as pd  #pandas provides various data structures and operations for manipulating numerical data and time series
import matplotlib.pyplot as plt  #matplotlib.pyplot is a plotting library used for 2D graphics
import seaborn as sns   #seaborn is a library for making statistical graphics in Python
```

```python
most_popular = df_tracks.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
#Popular sings which have popularity greater than 90
#inplace function is used so as to not change the original dataframe
```

```python
df_tracks.set_index("release_date", inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
#Changing of the index by using inplace=true
#release date should have a format of date
```

```python
df_tracks[["artists"]].iloc[18]
#Filtering specific info from dataset
#iloc is index location
```

```python
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace=True, axis=1)
#convert the duration (in ms) to seconds by using lambda function and dividing by 1000
```

```python
df_tracks.duration.head()
```

```python
#visualization that would a coorelation map
#dropping three unwanted columns and then apply pearson correlation method
corr_df=df_tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
#heatmap function of seaborn library is used to create the correlation map
#annottation is set to be true so that is write the data value in each set
#fmt=".1g" - it is a string formatting code to use, when adding annotation
#vmin and vmax are used to anchor the colour map , cmap - colour map
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="Black")
heatmap.set_title("Correlation HeatMap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
#sampling the data
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df)) #0.4% of original dataset
```

```python
#Regression plot between loudness and Energy
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df,y = "loudness", x = "energy", color = "c").set(title = "Loudness vs Energy Correlation")
```

```python
#Regression plot between Popularity and Acousticness
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df,y = "popularity", x = "acousticness", color = "b").set(title = "Popularity vs Acousticness Correlation")
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
#Creating a new column - year from our release_date column
```

```python
#Distribution plot to visualize the total no of songs in each year since 1992
#kind = hist stands for histogram
sns.displot(years,discrete=True,aspect=2,height=5, kind="hist").set(title="Number of songs per year")
```

```python
#Duration of songs over the years
#polting the bar plot
total_dr = df_tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title="Year vs Duration")
plt.xticks(rotation=90)
```
